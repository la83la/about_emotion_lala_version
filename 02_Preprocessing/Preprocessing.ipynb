```python
import pandas as pd
import numpy as np
import os
```

```python
data_list = os.listdir()
fb_list = data_list.copy()
news_list = data_list.copy()

# News list
unwanted = ['.ipynb_checkpoints', 'Preprocessing.ipynb', '2022-11-25_Facebook_Post_Data.pkl',\
                              '2022-12-01_Facebook_Post_Data.pkl',  '2022-12-08_Facebook_Post_Data.pkl']
news_list = [ele for ele in news_list if ele not in unwanted]
# fb list
unwanted = ['.ipynb_checkpoints', 'Preprocessing.ipynb','2022-10-29_NewsData.pkl',
 '2022-11-04_NewsData.pkl',
 '2022-11-09_NewsData.pkl',
 '2022-11-16_NewsData.pkl',
 '2022-11-24_NewsData.pkl',
 '2022-12-01_NewsData.pkl',
 '2022-12-08_NewsData.pkl']

fb_list = [ele for ele in fb_list if ele not in unwanted]
```

```python
# fb
fb_combine = []
for dataset in fb_list:
    tmp = pd.read_pickle(dataset)
    fb_combine.append(tmp)
fb_merge = pd.concat(fb_combine).reset_index(drop=True)
```

```python
# news
news_combine = []
for dataset in news_list:
    tmp = pd.read_pickle(dataset)
    news_combine.append(tmp)
news_merge = pd.concat(news_combine).reset_index(drop=True)
```
