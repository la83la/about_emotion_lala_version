```python
def find_word(txt):
    word = ""
    start = txt.find("word1") + 6
    while not txt[start].isspace():
        word += txt[start]
        start += 1
    return word
```

# Count the sentiment score 

```python
def count_negative_point(post):
    neg_score = 0
    for sn in strong_negative_trans: 
        if sn in post: 
            neg_score += 1.5
    for wn in weak_negative_trans: 
        if wn in post: 
            neg_score += 1
    return neg_score

def count_positive_point(post):
    pos_score = 0
    for sp in strong_positive_trans: 
        if sp in post: 
            pos_score += 1.5
    for wp in weak_positive_trans: 
        if wp in post: 
            pos_score += 1
    return pos_scores
```

```python
score_df = data_extract.copy()
score_df['positive_point'] = score_df['tokens'].map(count_positive_point)
score_df['negative_point'] = score_df['tokens'].map(count_negative_point)
```

```python
score_df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Step1: Pos. - Neg.
tmp = score_df.copy()
tmp['OF_Val'] = tmp['positive_point'] - tmp['negative_point']

# Step2: minMax scale
scaler = MinMaxScaler()
tmp['OF_Scaled'] = scaler.fit_transform(tmp['OF_Val'].values.reshape(-1,1))

# Step3: Linear transform
tmp['OF_Transform'] = tmp['OF_Scaled'].apply(lambda x: x*2-1)
tmp = tmp.drop(columns=['positive_point', 'negative_point', 'OF_Val', 'OF_Scaled'])
tmp
```

```python
agg_data = clean_data.copy()
agg_data['date_prune'] = agg_data.apply(lambda x: datetime.strftime(x['datetime'], "%Y-%m-%d"),axis=1)
del agg_data['datetime'] 
agg_data.head()
```

```python
sector = agg_data.groupby("date_prune")
sector.size()
```

```python
condense_df = agg_data.drop(columns=['content', 'tokens'])
condense_df = condense_df.groupby("date_prune").agg({'OF_Transform':'mean'})
condense_df['date'] = condense_df.index
condense_df = condense_df.reset_index(drop=True)
condense_df['date'] = pd.to_datetime(condense_df["date"])
condense_df.head()
```
